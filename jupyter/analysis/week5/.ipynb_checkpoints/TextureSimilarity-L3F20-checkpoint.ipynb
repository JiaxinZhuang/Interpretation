## Import module

```python
import os
os.environ["CUDA_DEVICE_ORDER"] = "PCI_BUS_ID"
os.environ["CUDA_VISIBLE_DEVICES"] = "0"
import sys
sys.path.append("../../../src/")
sys.path.append("../..")

import torch
import matplotlib.pyplot as plt
import matplotlib as mpl
mpl.rcParams['figure.dpi'] = 600
import numpy as np
from torchvision import transforms
from PIL import Image

from aux.utils import zscore, get_DISTS, get_top_k
from aux.visualization import plot_cov_matrix
```

# Load Brodatz

```python
data_dir = "/media/lincolnzjx/HardDisk/Datasets/Brodatz/"
```

```python
width = height =  224
mean = [0.485, 0.456, 0.406]
std = [0.229, 0.224, 0.225]
device = torch.device("cuda:0" if torch.cuda.is_available() else "cpu")
data = []
data_paths = []


for index, path in enumerate(os.listdir(data_dir)):
    file_path = os.path.join(data_dir, path)
    # print(index, file_path)
    img = Image.open(file_path).convert("RGB").resize((width, height))
    img = np.array(img).astype("float32")
    data.append(img)
    data_paths.append(file_path)
    
data = np.array(data)
# ref_imgs = zscore(data, mean, std)
ref_imgs  = torch.tensor(data) / 255.0
ref_imgs = ref_imgs.permute((0, 3, 1, 2))
print(ref_imgs.size())
```

```python
imgs = ref_imgs[0].unsqueeze(0)
print(imgs.size())
```

# DISTS

```python
cov_matrix = plot_cov_matrix(ref_imgs, method="DISTS", device=device)
```

```python
rows = cov_matrix.shape[0]
for index in range(rows):
    print("Img-{} Similar Index: {},\nDistance: {}".format(
         index, *get_top_k(cov_matrix, 
                   filter_index=index, top_k=3, descend=True)))
```
