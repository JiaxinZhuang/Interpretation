## Hyperparameter

```python
resume = "037-0"
model_dir = "/home/lincolnzjx/Desktop/Interpretation/saved/models"
generated_dir = "/home/lincolnzjx/Desktop/saved/generated/"
backbone = "vgg16"

device = torch.device("cuda:0" if torch.cuda.is_available() else "cpu")
```

## Load modal from dict

```python
# model
net = model.Network(backbone=backbone, num_classes=1000)
net.to(device)

# resume from model
resume_exp = resume.split("-")[0]
resume_epoch = resume.split("-")[1]
print("Resume from model from exp: {} at epoch {}".format(resume_exp, resume_epoch))
resume_path = os.path.join(model_dir, str(resume_exp), str(resume_epoch))
ckpt = torch.load(resume_path, map_location=device)
net.load_state_dict(ckpt) 

# Class & data
batch_size = 256
num_workers = 4
mean = [0.485, 0.456, 0.406]
std = [0.229, 0.224, 0.225]
reverse_mean = [-0.485, -0.456, -0.406]
reverse_std = [1/0.229, 1/0.224, 1/0.225]
```

## Load Original Data 

```python
class_y_true_y_pred_acc = {}
for class_index in range(1000):
    # Load data
    train_transform = transforms.Compose([                                                                                                                                
        transforms.Resize((224, 224), interpolation=Image.BILINEAR),                                                                                              
        transforms.ToTensor(),                                                                                                                                            
        transforms.Normalize(mean, std)                                                                                                                                   
           ])         
    trainset = imagenet.ImageNet(root="/media/lincolnzjx/HardDisk/Datasets/", 
                                 is_train=True, transform=train_transform)
    len_specific_category = len(trainset.targets_imgs_dict[class_index])
    print(len_specific_category)
    trainset.set_data([class_index], len_specific_category)
    trainloader = torch.utils.data.DataLoader(trainset,
                                              batch_size=batch_size,
                                              shuffle=False,
                                              num_workers=num_workers,
                                              pin_memory=True)
    y_true, y_pred, acc = get_gt_pred_specific_class(net,
                                                     trainloader,
                                                     order=[0,1,2],
                                                     device=device
                                                     )
    print("class_index:{}, Acc1 {}".format(class_index, acc))
    class_y_true_y_pred_acc[class_index] = [y_true, y_pred, acc]
```

```python
RGB_gt_pred_acc = "./rgb_gt_pred_acc.pickle"
with open(RGB_gt_pred_acc, "wb") as handle:
    pickle.dump(class_y_true_y_pred_acc, handle)
```

## Load Gray Image

```python
gray_class_y_true_y_pred_acc = {}
for class_index in range(1000):
    # Load data
    train_transform = transforms.Compose([                                                                                                                                
        transforms.Resize((224, 224), interpolation=Image.BILINEAR),                                                                                              
        transforms.Grayscale(num_output_channels=3),
        transforms.ToTensor(),                                                                                                                                            
        transforms.Normalize(mean, std)                                                                                                                                   
       ])         
    trainset = imagenet.ImageNet(root="/media/lincolnzjx/HardDisk/Datasets/", 
                                 is_train=True, transform=train_transform)
    len_specific_category = len(trainset.targets_imgs_dict[class_index])
    print(len_specific_category)
    trainset.set_data([class_index], len_specific_category)
    trainloader = torch.utils.data.DataLoader(trainset,
                                              batch_size=batch_size,
                                              shuffle=False,
                                              num_workers=num_workers,
                                              pin_memory=True)
    y_true, y_pred, acc = get_gt_pred_specific_class(trainloader)
    print("class_index:{}, Acc1 {}".format(class_index, acc))
    gray_class_y_true_y_pred_acc[class_index] = [y_true, y_pred, acc]
```

```python
gray_gt_pred_acc = "./gray_pred_acc.pickle"
with open(gray_gt_pred_acc, "wb") as handle:
    pickle.dump(gray_class_y_true_y_pred_acc, handle)
```
