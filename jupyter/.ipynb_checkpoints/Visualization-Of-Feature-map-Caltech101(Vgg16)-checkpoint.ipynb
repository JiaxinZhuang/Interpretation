```python
def visualize_features_map(img_index, layer_index, features_map, opt_features_map, cols=4):
    """Visulize feature map.
    """
    features_map = copy.deepcopy(features_map)[layer_index]
    opt_features_map = copy.deepcopy(opt_features_map)[layer_index]
    assert features_map.shape ==  opt_features_map.shape
    features_map = features_map.transpose((0, 2, 3, 1))
    opt_features_map = opt_features_map.transpose((0, 2, 3, 1))
    n_filters = features_map.shape[-1]
    rows = math.ceil(n_filters / cols)
    
    fig = plt.figure(constrained_layout=True)
    gs = GridSpec(rows, cols*2, figure=fig)

#     font = {'family' : 'normal',
#              'weight' : 'bold',
#              'size'   : 8}
    width = height = 224
    index = 0
    for row in range(0, rows):
        for col in range(0, cols):
            # specify subplot and turn off axis
            ax = fig.add_subplot(gs[row, col*2: col*2+2])
            ax.title.set_text("{}".format(index-1))
            # ax.title.set_fontsize("xx-small")
            ax.set_xticks([])
            ax.set_yticks([])
            # plot feature maps in grayscale
            img = Image.fromarray(features_map[img_index, :, :, index]).resize((width, height), PIL.Image.BICUBIC)
            opt_img = Image.fromarray(opt_features_map[img_index, :, :, index]).resize((width, height), PIL.Image.BICUBIC)
            cat_img = cat_img_hrizontal(img, opt_img)
            #img = img.resize((224, 224), resample=PIL.Image.BICUBIC)
            cat_img_np = np.array(cat_img)
            plt.imshow(cat_img_np, cmap="gray")
            ax.set_title("{} channel".format(index))
            #plt.subplots_adjust(hspace=0.6, wspace=0)
            index += 1
    # show figure
    fig.suptitle("Layer-{}".format(layer_index))
    pyplot.show()
```

```python
def cat_img_hrizontal(img_left, img_right):
    """Concat image horizontal.
    """
    width, height = img_left.size
    out = Image.new("F", (width*2, height))
    out.paste(img_left, (0, 0))
    out.paste(img_right, (width, 0))
    return out
```
